```python
import os
import pandas as pd
import glob
```

```python
os.listdir(r".\res_gossiping")
```

```python
# unix style
glob.glob(r".\res_gossiping\*.txt")
```

```python
data = []

for file_path in glob.glob(r".\res_gossiping\*.txt"):
    with open(file_path, "r", encoding="UTF-8") as f:
        # Read the whole article as string
        article_content = f.read()
        reply_str = article_content.split("---split---")[1]
        row_data = reply_str.split("\n")[1:-1]
        # row_data = [x.split(":",1)[1] for x in row_data]
        data.append(row_data)
        # print(data)

df = pd.DataFrame(
    data=data,
    columns=["推", "噓", "分數", "作者", "標題", "時間"],
)
df
```

```python
# save as csv flie
df.to_csv("ptt.csv", index=False)
df.to_csv(
    "ptt_nocolumn.csv",
    index=False,
    header=None,
)
```

```python
"""
"推: 17" -> 17
"""


def column_transfomation(input_str: str) -> int:
    return int(
        input_str.split(": ")[1]
    )
```

```python
df["推"] = df["推"].apply(column_transfomation)
df
```
